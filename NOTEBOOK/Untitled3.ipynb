```python
#%matplotlib notebook
%matplotlib inline
import time
import pylab as pl
import matplotlib.pyplot as plt
from IPython import display

for i in range(20):
    plt.clf()
    plt.plot(pl.randn(100))
    display.clear_output(wait=True)
    display.display(plt.gcf)
    time.sleep(0.1)
```

```python
while True:
    try:
        plt.clf()
        plt.plot(pl.randn(100))
        display.display(plt.gcf())
        display.clear_output(wait=True)
        time.sleep(0.1) 
    except KeyboardInterrupt:
        break
```
